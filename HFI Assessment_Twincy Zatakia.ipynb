```python
import pandas as pd
import numpy as np
import nltk
from nltk import word_tokenize
import string
```

```python
df = pd.read_csv("C:/Users/dhruv/Desktop/HFI_assessment/assignment_data.csv")
```

```python
df.head()
```

```python
df['Summary'] = df['Adverse Reactions'].str.lower() # preprocessing data
```

```python
df['Summary'] = df['Summary'].apply(lambda x:''.join([i for i in x                # remove punctuation
                                                  if i not in string.punctuation]))
```

```python
import gensim                                             # remove stopwords
from gensim.parsing.preprocessing import remove_stopwords

clean = []
for i in df['Summary']:
    clean.append(remove_stopwords(i))
```

```python
df['Summary']=clean
```

```python
nltk.pos_tag(word_tokenize(df['Summary'][0]))   # Extracting Parts of speech and tokenization
```

```python
def tagging(content):                                   
    temp = []
    for i in content:
        temp.append(nltk.pos_tag(word_tokenize(i)))
    return temp
```

```python
postag = tagging(df['Summary'])
```

```python
para_data = []                                   # Extracting only particular set of words 
result = []
for i in range(len(postag)):
    for j in range(len(postag[i])):
        if postag[i][j][1]=='NN':
            para_data.append(postag[i][j][0])
    result.append(para_data)
    para_data=[]
```

```python
df['PreProcessed']=result
```

```python
df['PreProcessed']
```

```python
df['Summary'] = result
```
